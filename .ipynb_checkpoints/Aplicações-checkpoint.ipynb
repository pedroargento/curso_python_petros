```python
import pandas as pd
import numpy as np
from scipy.stats import norm

```

```python
def BSM(S0, exp_date, vol, strike, rf = 0.05, tipo = 'call'):
    '''
    S0 -> Preço atual do ativo
    exp_date -> tempo até vencimento
    vol -> volatilidade do ativo
    strike -> preço acordado de strike da opção
    rf -> taxa livre de risco
    tipo -> valores: 'call' -> opção de compra; 'put' -> opção de venda
    '''
    
    d1 = (np.log(S0/strike) + rf + exp_date*vol**2/2)/(vol*np.sqrt(exp_date))
    d2 = d1 - vol*np.sqrt(exp_date)
    
    if tipo == 'call':
        return S0*norm.cdf(d1) - strike*np.exp(-rf*exp_date)*norm.cdf(d2, 0, 1)
    if tipo == 'put':
        return strike*np.exp(-rf*exp_date)*norm.cdf(-d2, 0, 1) - S0*norm.cdf(-d1)
    else:
        print('tipo de opção não reconhecido')
        return 0
```

```python
#PUT S&P500
BSM(3427.18, 0.11, 0.255, 3400, 0.0025, tipo = 'call')
```

```python
def MGB (sigma, steps_per_year, n_years, s_0=100, mu = 0, n_scenarios = 10000):
    dt = 1/steps_per_year
    n_steps = int(n_years*steps_per_year) + 1
    rets_plus_1 = np.random.normal(loc=(1+mu)**dt, scale=(sigma*np.sqrt(dt)), size=(n_steps, n_scenarios))
    rets_plus_1[0] = 1
    prices = s_0*pd.DataFrame(rets_plus_1).cumprod()
    return prices
```

```python
def port_return(weights, expected_returns):
    return weights.T@expected_returns
```

```python
def port_vol(weights, cov_matrix):
    return (weights.T@cov_matrix@weights)**(1/2)
```
