```python
def getData (symbol):
    r = requests.get('https://finance.yahoo.com/quote/'+symbol) 
    b = BeautifulSoup(r.text, 'html.parser') 
    b.findAll('span')
    quote_summary = b.find('div', attrs={'id':"quote-summary"}) 
    data = quote_summary.findAll('span')
    data = [i.text for i in data]
    return {data[i]: data[i + 1] for i in range(0,29,2)} 
```

```python
getData('AAPL')
```

```python
symbols = ['AAPL', 'SPY', 'AMZN', 'TSLA']
```

```python
for s in symbols:
    print(getData(s))
```

```python
import pandas as pd
df = pd.DataFrame(list(map(lambda x: getData(x), symbols)))
```

```python
df.index = symbols
```
